## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия формируется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае индивидуальная оценка участника становится равной общей оценке команды за проект (т.е. в идеальном случае — вторая оценка равна первой оценке).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(04032026)`.

На основании круговой диаграммы можно сделать вывод, что Netflix активно развивает контент, ориентированный на различные возрастные и целевые аудитории\. Преобладающие рейтинговые группы, такие как TV\-14 \(21\.2%\), TV\-MA \(16\.4%\) и PG \(15\.2%\), показывают, что платформа нацелена как на взрослую аудиторию, так и на молодежь и семейных зрителей\. Это свидетельствует о значительном количестве контента для подростков и взрослых \(рейтинг TV\-14 и TV\-MA\), а также об активном продвижении контента для детей и семейных шоу \(рейтинг PG\)\. Таким образом, Netflix стремится создать разнообразный контент, который удовлетворяет потребностям разных групп зрителей, что позволяет привлекать и удерживать пользователей с разными предпочтениями и возрастными категориями\.

В США действуют две параллельные системы рейтингов, одна для кино, другая для телевизионного контента\. Важно понимать что некоторые из них являются прямыми аналогами друг друга, просто применяются к разным типам медиа\.

Телевизионные рейтинги \(TV\) — для сериалов и ТВ\-шоу:
TV\-Y — специально для детей любого возраста, полностью безопасный контент
TV\-Y7 — для детей от 7 лет, может содержать лёгкие конфликтные ситуации
TV\-Y7\-FV — для детей от 7 лет, но допускает фантастическое насилие \(Fantasy Violence\)
TV\-G — для всей семьи, без возрастных ограничений
TV\-PG — родительский контроль желателен, может быть лёгкий юмор или напряжённые сцены
TV\-14 — не рекомендуется детям до 14 лет, содержит взрослые темы
TV\-MA — только для взрослой аудитории, может содержать откровенные сцены и насилие
Кинорейтинги \(MPAA\) — для полнометражных фильмов:
G \- для всех возрастов, без ограничений
PG — родительский контроль желателен, может быть грубый юмор или лёгкое насилие
PG\-13 — не рекомендуется детям до 13 лет
R — только для взрослых или в сопровождении родителей
Без рейтинга:
NR — контент не проходил официальную рейтинговую оценку
UR — Unrated, версия фильма или шоу без присвоенного рейтинга

Ключевой момент для анализа: TV\-G и G это один и тот же возрастной ценз, как и TV\-PG и PG, TV\-14 и PG\-13, TV\-MA и R\. Разница только в типе контента телевизионный или кинематографический\. Это важно учитывать при сравнении рейтинговых групп между собой\.

Для этого создадим еще один столбец с объединенными рейтингами для ТВ\-контента и кино

```
rating_map = {'TV-G': 'G', 'TV-PG': 'PG', 'TV-14': 'PG-13', 'TV-MA': 'R', 'TV-Y': 'Y', 'TV-Y7': 'Y7', 'TV-Y7-FV': 'Y7-FV'}
```

```
data['rating_overall'] = data['rating'].replace(rating_map)
data.head()
```

```
rating_and_levels = data.groupby('rating_overall')['ratingLevel'].nunique(dropna=False).sort_values(ascending=False)
rating_and_levels
```

```
ax = rating_and_levels.plot(kind='bar', figsize=(16, 9))
ax.bar_label(ax.containers[0])
plt.title('Количество уникальных ratingLevel для каждого rating')
plt.xlabel('rating')
plt.ylabel('Число уникальных ratingLevel')
plt.xticks(rotation=45)
plt.show()
```

По итогу видим, что наибольшее число уникальных ratingLevel имеет категория PG, поскольку PG \- категория, которая предусматривает, что взрослый читает описание шоу/фильма/сериала и решает, хочет ли он, чтобы это видел его ребенок или нет\. При объединении нашего датасета с отчётом Netflix Engagement Report мы столкнулись с типичной проблемой — названия одного и того же контента записаны по\-разному\. Например, в наших данных сериал называется просто "Breaking Bad", а в отчёте он разбит на сезоны: "Breaking Bad: Season 1", "Breaking Bad: Season 2" и так далее\. Просто сравнить строки тут не получится, поэтому пришлось писать свой алгоритм поиска\.

Сначала мы собрали все названия из нашего датасета в отдельное множество, привели их к нижнему регистру и убрали лишние пробелы — это стандартная чистка, чтобы потом не спотыкаться на мелочах вроде лишнего пробела в конце или заглавной буквы\.

Дальше загрузили Engagement Report и сразу наткнулись на проблему с колонкой часов просмотра — там числа хранились как текст с запятыми, типа "812,100,000"\. Пришлось убрать все запятые и перевести в нормальный числовой формат\. Ещё удалили строки с пустыми названиями, иначе на них всё падало\.

Самая интересная часть — сам поиск совпадений\. Для каждого названия из нашего датасета мы строили регулярное выражение: сначала идёт само название, а после него допускаются только четыре варианта окончания — двоеточие \(как в "Breaking Bad: Season 1"\), скобка \(как в "Love \(2015\)"\), двойной слэш \(в отчёте встречалось "// Pieles"\) или просто конец строки\. Это было ключевым решением: например, по слову "Love" мы находили "Love: Season 1" и "Love \(2015\)", но не находили "Love is Blind", потому что после "love" там идёт пробел и следующее слово, а это не входит в наши разрешённые варианты\. Так мы отсекли ложные срабатывания и не суммировали просмотры совершенно разных шоу в одну цифру\.

Дальше через \`str\.match\` проверяли каждую строку отчёта и собирали все совпадения\. Если находили хоть одно — суммировали часы просмотра по всем найденным сезонам \(например, для Breaking Bad просто складывали все сезоны в общую кучу\) и сохраняли результат\.

В конце превратили всё в нормальную таблицу, приджойнили к основному датасету и добавили фичу — поделили пользовательские оценки на 10, чтобы привести их к одной шкале с IMDb и можно было нормально сравнивать\.
